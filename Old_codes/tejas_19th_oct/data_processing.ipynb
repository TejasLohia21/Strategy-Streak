```python
import yfinance as yf
import time

stock_symbols = ["RELIANCE.NS", "TCS.NS", "HDFCBANK.NS"]  # NSE symbols for yfinance
while True:
    for symbol in stock_symbols:
        try:
            stock = yf.Ticker(symbol)
            data = stock.history(period="1d", interval="5m")
            last_price = data['Close'].iloc[-1]  # Get the last closing price
            print(f"{symbol}: Last Price = {last_price}")
        except Exception as e:
            print(f"Error fetching data for {symbol}: {e}")
    time.sleep(30)  # Wait for 5 minutes
```

```python
import yfinance as yf
import pandas as pd

# Define the stock symbol for Reliance
symbol = "RELIANCE.NS"

# Fetch historical data for yesterday at 1-minute intervals
historical_data = yf.download(symbol, period='1d', interval='1m')

# Filter the data for yesterday
yesterday_data = historical_data[historical_data.index.date == (pd.Timestamp.now() - pd.Timedelta(days=1)).date()]

# Display the data
print("Historical Data for Reliance for Yesterday (1-minute intervals):")
print(yesterday_data)

# Optional: Save to a CSV file
yesterday_data.to_csv(f'{symbol}_historical_data_yesterday.csv')
```
